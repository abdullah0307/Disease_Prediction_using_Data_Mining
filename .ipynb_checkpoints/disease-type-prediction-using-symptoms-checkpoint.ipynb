**Importing libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix
import seaborn as sns
```

**Read the dataset**

```python
df = pd.read_csv('disease_dataset.csv')
df.head(10)
```

**Dataset info**

```python
df.info()
```

**Dataset characteristics**

```python
df.describe()
```

**Check for null and NaN values**

```python
df.isna().sum()
```

```python
df.isnull().sum()
```

**Remove the trailing space from the symptom columns**

```python
cols = df.columns
data = df[cols].values.flatten()

s = pd.Series(data)
s = s.str.strip()
s = s.values.reshape(df.shape)

df = pd.DataFrame(s, columns=df.columns)
df.head()
```

**Fill the NaN values with zero**

```python
df = df.fillna(0)
df.head()
```

**Encode symptoms in the data with the symptom rank**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for column in df.columns:
    df[column] = le.fit_transform(df[column].astype(str))
```

**Plot the dataset attribute with each other**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,45))

plot_count = 1
for feature in list(df.columns):
    plt.subplot(11,3,plot_count)
    plt.scatter(df[feature], df['Disease'])
    plt.xlabel(feature.replace('_',' ').title())
    plt.ylabel('Disease')
    plot_count+=1
    
plt.show()
```

**Check if entire columns have zero values so we can drop those values**

```python
(df[cols] == 0).all()
```

```python
df['Disease'].value_counts()
```

**Get the names of diseases from data**

```python
df['Disease'].unique()
```

**Select the features as symptoms column and label as Disease column**

```python
data = df.iloc[:,1:].values
labels = df['Disease'].values
```

**Split the data, labels with shuffle**

```python
x_train, x_test, y_train, y_test = train_test_split(data, labels, shuffle=True, train_size = 0.85)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

**Initialize and train a Support vector classifier**

```python
model = SVC()
model.fit(x_train, y_train)
```

**Throw predictions**

```python
preds = model.predict(x_test)
```

**Plot the confusion matrix for 25 diseases and calculate f1, accuracy**

```python
conf_mat = confusion_matrix(y_test, preds)
df_cm = pd.DataFrame(conf_mat, index=df['Disease'].unique(), columns=df['Disease'].unique())
print('F1-score% =', f1_score(y_test, preds, average='macro')*100, '|', 'Accuracy% =', accuracy_score(y_test, preds)*100)
sns.heatmap(df_cm)
```

**Reading dataset of disease**

```python
# read the dataset
import pandas as pd
import numpy as np
# let assume the patient have the following symptoms
df = pd.read_csv("disease_dataset.csv")
```

**remove all the spaces**

```python
cols = df.columns
data = df[cols].values.flatten()

s = pd.Series(data)
s = s.str.strip()
s = s.values.reshape(df.shape)

df = pd.DataFrame(s, columns=df.columns)
```

```python
# predict the disease for the following symptoms
symptoms = ['itching','skin_rash','stomach_pain','burning_micturition','spotting_urination']
test = df.copy()
for symptom in symptoms:
    for column in test.columns:
        if not test[test[column] == symptom].empty:
            test = test[test[column] == symptom]
test['Disease'].unique()
```

```python
# Calculate the disease by inputting by the option
test = df.copy()
for symptom in test.columns[1:]:
    # if the options are less than or equal to one 
    if len(test[symptom].unique()) <= 1:
        break
        
    sy = input(test[symptom].unique())
    
    for column in test.columns:
        if not test[test[column] == sy].empty:
            test = test[test[column] == sy]
            print(test)
            
print(test['Disease'].unique()[0])
```

```python
# find all the unique symptoms of the dataset
unique = []
for key in df.columns:
    for i in df[key]:
        if not i in unique:
            unique.append(i)
```

```python
test = pd.DataFrame(index=[], columns=[i for i in range(18)])
test.columns = df.columns

for i in range(len(df.columns)):
    sy = input(df[df.columns[i]].unique())
    if df[df.iloc[:,i] == sy].empty:
        continue
    test = pd.concat([test, df[df.iloc[:,i] == 'cough']], ignore_index=True)
test
```

```python
# Calculate the disease by inputting by the option improved
test = df.copy()

# empty dataframe
out = pd.DataFrame(index=[], columns=df.columns)

# options selected
options_selected = []

# loop to travers from all the columns for symptoms
for symptom in out.columns[1:]:
    
    #options to show
    options = [] 
    
    # Extract the options from the column
    for i in test[symptom].unique():
        if not i in options_selected and i != np.nan:
            options.append(i)
            
    # if the options are less than or equal to one move to next column
    if len(options) <= 1:
        continue
       
    # choose the options for the input 
    sy = input(options)
    options_selected.append(sy)
    
    # Reduce the test df according to the symptoms
    for i in out.columns[1:]:
        if test[test[i]==sy].empty:
            continue
        test = test[test[i] == sy]
        out = pd.concat([out, test], ignore_index=True)
    out = test.copy()
    print(out)
    
print(out['Disease'].unique()[0])
```
